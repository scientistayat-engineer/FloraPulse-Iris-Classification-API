# 📊 Step 2: Ingest Dataset & Establish Validation Matrices

```
# ==============================================================================
# CELL 2: DATA LOADING AND CROSS-VALIDATION SPLIT
# ==============================================================================
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Load the authentic structural Iris data layer from Scikit-Learn
iris_dataset = load_iris()
X_features = iris_dataset.data
y_targets = iris_dataset.target

feature_dimensions = iris_dataset.feature_names
categorical_classes = iris_dataset.target_names

print(f"Target Species Array: {list(categorical_classes)}")
print(f"Feature Space Matrix: {feature_dimensions}")

# Perform high-integrity split (80% training matrix, 20% validation pool)
X_train, X_test, y_train, y_test = train_test_split(
    X_features, y_targets, test_size=0.2, random_state=42, stratify=y_targets
)
print(f"Split Matrices Verified: Train Shape {X_train.shape} | Test Shape {X_test.shape}")
```

# 🧠 Step 3: Model Architecture Optimization & Verification

```
# ==============================================================================
# CELL 3: MODEL ESTIMATOR TRAINING AND PERFORMANCE AUDIT
# ==============================================================================
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Initialize an enterprise-grade Random Forest Classifier
model_estimator = RandomForestClassifier(n_estimators=100, random_state=42)
model_estimator.fit(X_train, y_train)

# Evaluate classification precision limits
test_predictions = model_estimator.predict(X_test)
calculated_accuracy = accuracy_score(y_test, test_predictions) * 100

print(f"🏆 System Analytics Matrix | Target Verification Accuracy: {calculated_accuracy:.2f}%\n")
print("--- Detailed Production Classification Report ---")
print(classification_report(y_test, test_predictions, target_names=categorical_classes))
```

# 💾 Step 4: Model Binary Layer Serialization

```
# ==============================================================================
# CELL 4: ARTIFACT SERIALIZATION STORAGE
# ==============================================================================
import joblib

# Serialize model state vector to local disk path
production_model_path = "iris_random_forest_model.pkl"
joblib.dump(model_estimator, production_model_path)

print(f"Artifact Serialized Successfully: '{production_model_path}' is ready for production API integration.")
```

# 🌐 Step 5: Programmatic Writing of the FastAPI Core Service File

```
# ==============================================================================
# CELL 5: PROGRAMMATIC SOURCE SCRIPT GENERATION FOR THE BACKEND APPLICATION
# ==============================================================================
fastapi_application_code = """
import joblib
import numpy as np
from fastapi import FastAPI
from pydantic import BaseModel, Field

# 1. Initialize production framework instance
app = FastAPI(
    title="FloraPulse - Iris Variant Inference Core",
    description="Automated biological classification engine exposing high-availability REST endpoints.",
    version="1.0.0"
)

# 2. Load the binary serialization layer safely
predictive_pipeline = joblib.load("iris_random_forest_model.pkl")
target_species_mapping = ["setosa", "versicolor", "virginica"]

# 3. Define robust data schema validation framework using Pydantic
class IrisInferenceSchema(BaseModel):
    sepal_length: float = Field(..., example=5.1, description="Length of the sepal in centimeters.")
    sepal_width: float = Field(..., example=3.5, description="Width of the sepal in centimeters.")
    petal_length: float = Field(..., example=1.4, description="Length of the petal in centimeters.")
    petal_width: float = Field(..., example=0.2, description="Width of the petal in centimeters.")

# 4. Endpoint 1: Health Diagnostic / Greetings Router Gateway
@app.get("/")
def system_health_handshake():
    return {
        "status": "online",
        "message": "Welcome to the Vanguard Systems Iris Species Classification API Gateway!"
    }

# 5. Endpoint 2: Automated Real-time Matrix Prediction
@app.post("/predict")
def route_live_inference(payload: IrisInferenceSchema):
    # Format structural input fields into standard numerical vector row
    feature_vector = np.array([[
        payload.sepal_length,
        payload.sepal_width,
        payload.petal_length,
        payload.petal_width
    ]])

    # Execute class extraction inference via random forest algorithm state
    class_index_result = predictive_pipeline.predict(feature_vector)[0]
    resolved_species_string = target_species_mapping[class_index_result]

    return {
        "prediction": resolved_species_string
    }
"""

with open("app.py", "w") as storage_writer:
    storage_writer.write(fastapi_application_code)
print("System Status: Production server core saved to 'app.py' successfully.")
```

# 🚀 Step 6: Start the Asynchronous Web Application Server

```
# ==============================================================================
# CELL 6: ASYNCHRONOUS SERVER INITIALIZATION
# ==============================================================================
import uvicorn
import multiprocessing

# Running the production FastAPI server file via Uvicorn in a separate background process
def initiate_gateway():
    uvicorn.run("app:app", host="127.0.0.1", port=8000, log_level="info")

server_daemon = multiprocessing.Process(target=initiate_gateway)
server_daemon.start()
print("Gateway Server Status: Asynchronous application background threading active on http://127.0.0.1:8000")
```

# 🧪 Step 7: Local Server Integration & Post Request Validation

```
# ==============================================================================
# CELL 7: SIMULATED ENDPOINT POST REQUEST VERIFICATION
# ==============================================================================
import time
import requests

# Give the background daemon a brief moment to initialize the routing sockets
time.sleep(3)

# 1. Validate Health Handshake Endpoint (GET)
health_check_response = requests.get("http://127.0.0.1:8000/")
print("GET [/] Response Matrix:", health_check_response.json())

# 2. Validate Automated Real-Time Model Inference Endpoint (POST)
test_payload_matrix = {
    "sepal_length": 5.1,
    "sepal_width": 3.5,
    "petal_length": 1.4,
    "petal_width": 0.2
}

inference_post_response = requests.post("http://127.0.0.1:8000/predict", json=test_payload_matrix)
print("\nPOST [/predict] Response Matrix:", inference_post_response.json())
```
